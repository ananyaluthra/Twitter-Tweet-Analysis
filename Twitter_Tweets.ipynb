```python
import pandas as pd
```

```python
tweets=pd.read_csv('/content/IRAhandle_tweets_1.csv')
```

```python
tweets=tweets.head(10000)
```

```python
english_tweets=tweets[tweets['language']=='English']
```

```python
english_tweets_noquestions=english_tweets[~english_tweets['content'].str.contains('\?',case=False)]
```

```python
english_tweets.to_csv('data_collection.tsv', sep="\t")
```

```python
english_tweets_noquestions
```

```python
english_tweets_noquestions['trump_mention']=''
```

```python
for a in english_tweets_noquestions['content']:
  d=english_tweets_noquestions.index.get_loc(english_tweets_noquestions.index[english_tweets_noquestions['content'] == a][0])
  if a.__contains__('Trump')==True:
    n=a.index('Trump')
    p=n+5
    y=str(a[n-1])
    if a.endswith('Trump') and y.isalnum()==False:
      english_tweets_noquestions.iloc[d,21]='T'
    else:
      x=str(a[p])
      y=str(a[n-1])
      if a.startswith('Trump') and x.isalnum()==False:
        english_tweets_noquestions.iloc[d,21]='T'
      elif x.isalnum()==False and y.isalnum()==False:
        english_tweets_noquestions.iloc[d,21]='T'
  else:
    english_tweets_noquestions.iloc[d,21]='F'
```

```python
english_tweets_noquestions['trump_mention']
```

```python
def trump_check(a):
  d=english_tweets_noquestions.index.get_loc(english_tweets_noquestions.index[english_tweets_noquestions['content'] == a][0])
  if a.__contains__('Trump')==True:
    n=a.index('Trump')
    p=n+5
    y=str(a[n-1])
    if a.endswith('Trump') and y.isalnum()==False:
      english_tweets_noquestions.iloc[d,21]='T'
    else:
      x=str(a[p])
      y=str(a[n-1])
      if a.startswith('Trump') and x.isalnum()==False:
        english_tweets_noquestions.iloc[d,21]='T'
      elif x.isalnum()==False and y.isalnum()==False:
        english_tweets_noquestions.iloc[d,21]='T'
  else:
    english_tweets_noquestions.iloc[d,21]='F'

```

```python
english_tweets_noquestions['content'].apply(trump_check)
```

```python
english_tweets_noquestions_trumpmention=english_tweets_noquestions[['tweet_id','publish_date','content','trump_mention']]
```

```python
english_tweets_noquestions_trumpmention.to_csv('dataset.tsv',sep='\t',header=True)
```
